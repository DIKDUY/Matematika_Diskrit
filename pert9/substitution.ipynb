```python
import string

class SubstitutionCipher:
    def __init__(self, key):
        self.key = key
        self.alphabet = string.ascii_lowercase
        self.cipher_dict = self._create_cipher_dict()
        self.decipher_dict = {v: k for k, v in self.cipher_dict.items()}
    
    def _create_cipher_dict(self):
        return {self.alphabet[i]: self.key[i] for i in range(len(self.alphabet))}
    
    def encrypt(self, plaintext):
        plaintext = plaintext.lower()
        ciphertext = ''.join(self.cipher_dict.get(char, char) for char in plaintext)
        return ciphertext
    
    def decrypt(self, ciphertext):
        ciphertext = ciphertext.lower()
        plaintext = ''.join(self.decipher_dict.get(char, char) for char in ciphertext)
        return plaintext

# Contoh penggunaan
key = "phqgiumeaylnofdxjkrcvstzwb"  # Kunci pengganti, harus 26 karakter unik
cipher = SubstitutionCipher(key)

plaintext = "hello world"
ciphertext = cipher.encrypt(plaintext)
print(f"Plaintext: {plaintext}")
print(f"Ciphertext: {ciphertext}")

decrypted_text = cipher.decrypt(ciphertext)
print(f"Decrypted text: {decrypted_text}")
```
